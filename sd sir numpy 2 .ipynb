# Numpy array copy vs view

- [**copy**] the copy should not be affected by changes made to the original array

- [**view**] the view should be affected by the changes made to the original array

```python
# creat copy

import numpy as np

arr=np.array([1,2,3,4,5])

x=arr.copy()
arr[0]=42

print(arr)
print(x)
```

```python
# create view
# the original array affect when view change

import numpy as np

ar=np.array([1,2,3,4,5])
x=ar.view()
ar[0]=42
x[2]=8

print(ar)
print("\n",x)
```

check if array owns its data

```python
import numpy as np

arr=np.array([1,2,3,4,5])

x=arr.copy()
y=arr.view()

print(x.base)  # copy owns the data, so it return none
print(y.base)  # view owns the data, so it returns the original array
```

# **iterating array using nditer()**
  - the function nditer() is a helping function that can be used from very basic to very advanced iteration

iterating on each scalar element

```python
# iterate through the following 3-d array:

import numpy as np

ar=np.array([[[1,2],[3,4]],
             [[5,6],[7,8]]])

for x in np.nditer(ar):
  print(x)
```

**Iterating array with different data type**
  - can use **op_dtypes argument** and pass it the expected data type to change the data type of element while iterating
  - numpy does not change the data type of the element in-place (where the elements is in array) so it needs some other space to perform this action , that extra space called **buffer**, and in order to enable it in **nditer() we pass flags=['buffered']**

```python
import numpy as np

arr=np.array([1,2,3])

for x in np.nditer(arr, flags=['buffered'], op_dtypes=['S']):
  print(x)

# string prefixed with b is a bytes object
```

```python
import numpy as np

arr=np.array([1,2,3])

for x in np.nditer(arr, flags=['buffered'], op_dtypes=['complex']):
  print(x)
```

# **joining arrays using stack function**

  - stack is the same as the concatenation, the **only difference is that stacking is done along a new axis**
  - concatenate two 1-d array along the second axis which result in putting them one over the other , ie.stacking
  - **stack() methond along with the axis. if axis is not explicity oassed it is taken as 0**

```python
import numpy as np

a=np.array([1,2,3])
b=np.array([5,6,7])

ab=np.stack((a,b),axis=1)
print(ab)
```

```python
import numpy as np

a=np.array([1,2,3])
b=np.array([4,5,6])

print(a)
print("\n",b)

print("\n----------by using stack (axis=1)-------------\n")
a_stack=np.stack((a,b),axis=1)
print(a_stack)

print("\n----------by using stack (axis=0)-------------\n")
b_stack=np.stack((a,b),axis=0)
print(b_stack)

print("\n----------by using concatenate (by default or axis=0)----------\n")
a_conct=np.concatenate((a,b))
print(a_conct)
```

# **Search sorted**

  - There is a method called searchsorted() which performs a binary search in the array, and **returns the index where the specified value would be inserted to maintain the search order**.
  - **The searchsorted() method is assumed to be used on sorted arrays**.

1.find the index where the value 7 would be inserted:

```python
import numpy as np

a=np.array([6,7,8,9])

x=np.searchsorted(a, 7)
print(x)
```

```python
import numpy as np

a=np.array([6,7,8,9])

x=np.searchsorted(a, 7)
print(x)
```

**Search From the Right Side**

  • By default the left most index is returned, but we can give side='right' to return the right most index instead.


1. Find the indexes where the value 7 should be inserted, starting from the right:

```python
import numpy as np

abc=np.array([6,7,8,9,10])

# Using the right side
# here the element that to be added is looked from the right side of the array
# And place according to the order
# And after that index number is given (as usual as it will only begin from the left side)

x=np.searchsorted(abc, 7, side="right")
print(x)
```

```python
import numpy as np

ar=np.array([6,7,8,9])

y=np.searchsorted(ar,10, side='right')
print(y)
```

- 2 is placed at index **2** because it should **come after 1 and before 5**.
- 4 is placed at index **4** because it should **come after the last element 3**.
- 6 is also placed at index **4**, as it would **come after the last element 3**.
- In summary, while **the searchsorted() function can still technically operate on an unsorted array, the results may not be meaningful in the context of maintaining order, as the function assumes the array is sorted**.

# **crating the filter array**

```python
import numpy as np

arr=np.array([41,42,43,44])

# create a empty list
filter_arr=[]

# go through each elements
for element in arr:
  if element > 42:
    filter_arr.append(True)
  else:
    filter_arr.append(False)

newarr=arr[filter_arr]

print("original array: ",arr)
print("\n boolean array: ",filter_arr)
print("\n filtered new array: ",newarr)
```

- Random number does NOT mean a different number every time. Random means something that can not be predicted logically.

**Pseudo Random and True Random**

- **Random numbers generated through a generation algorithm are called pseudo random**.

# **Random Permutations**



**Random Permutations of Elements**

  - A **permutation refers to an arrangement of elements. e.g. [3, 2, 1] is a permutation of [1, 2, 3] and vice-versa.**
  - The NumPy Random module provides two methods for this: **shuffle() and permutation().**

## **Shuffling Arrays**
  - Shuffle means changing arrangement of elements in-place. i.e. in the array itself.
  
**The shuffle() method makes changes to the original array.**

```python
from numpy import random
import numpy as np

ar=np.array([1,2,3,4,5])

random.shuffle(ar)
print(ar)
```

**Genarating permutation os arrays**

  - The permutation() method returns a re-arranged array (and leaves the original array un-changed).
